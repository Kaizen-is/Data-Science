###DataScience and AI Practicum
- Changing indexes

```python
import pandas as pd
city = {'Name':['Vien', 'NewYork', 'Tashkent', 'Namangan', 'Parice'],
        'Year':[2010, 2015, 2016, 2017, 2018],
        'Quantity':[56000, 34000, 23000, 65000, 87000],
        'Temp':[1.56, 1.45, 1.45, 1.53, 1.87]}

df = pd.DataFrame(city)
df
```

```python
df.index = ([1, 2,3 ,4 , 5])
df
```

`Reindexing`

```python
newindex = [4, 3, 2, 5, 1]
newindex.sort()
df.reindex(newindex)
```

```python
df.reindex([1, 2, 3])
```

```python
df.reindex(index = range(1, 11))
```

```python
df.reindex(index = range(1, 11), method = 'nearest')
```

```python
df.reindex(index = range(1, 11), method = 'ffill')
```

```python
df.reindex(columns=['Year', 'Temp', 'Name', 'Quantity'])
```

```python
df.reindex(columns = ['Name', 'Year', 'Quantity'])
```

```python
df.reindex(columns = ['Name', 'Year', 'Quantity'])
df.sample(3) # we get random lines
```
