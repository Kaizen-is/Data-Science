### Practice `.merge()`

`1.` 	Merge task1_users.csv and task2_scores.csv on id (simple inner join)

```python
import pandas as pd
```

```python
task1 = pd.read_csv('task1_users.csv')
task2 = pd.read_csv('task2_scores.csv')
print(task1, "\n------------------------------")
print(task2)
```

```python
user_score = pd.merge(task1, task2, on = 'id', how = 'inner')
user_score = user_score.loc[:, ['id', 'name', 'score']]
user_score.set_index('id', inplace = True)
user_score
```

`9.` Join 3 files (task10_users.csv, task10_products.csv, task10_countries.csv) to get full user/product/country info

```python
task10_users = pd.DataFrame({
    'user_id': [101, 102, 103],
    'product_id': ['A1', 'B2', 'C3']
}).set_index('user_id')

task10_products = pd.DataFrame({
    'product_id': ['A1', 'B2'],
    'price': [500, 700]
}).set_index('product_id')

task10_countries = pd.DataFrame({
    'user_id': [101, 102],
    'country': ['USA', 'Germany']
}).set_index('user_id')

print(task10_countries, '\n--------------------------------')
print(task10_products, '\n--------------------------------')
print(task10_users)
```

```python
user_coun = pd.merge(task10_users, task10_countries, on = 'user_id', how = 'inner')
user_prod_coun = pd.merge(user_coun, task10_products, on = 'product_id', how = 'inner')
user_coun
```
