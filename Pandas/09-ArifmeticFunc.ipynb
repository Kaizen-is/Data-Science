```python
import pandas as pd
import numpy as np

df1 = pd.DataFrame(np.arange(9.).reshape(3, 3), columns = list('abc'), index = ['Apple', 'Coconat', 'Melon'])
df2 = pd.DataFrame(np.arange(12.).reshape(4, 3), columns = list('adb'), index = ['Apple', 'Mango', 'Coconat', 'WatterMelon'])
```

```python
df1
```

```python
df2
```

```python
df1 + df2
```

```python
df1 - df2
```

```python
df1.add(df2) # to df1 is added df2
```

```python
df2.radd(df1)# radd, adds to the df in the case
```

```python
df1.add(df2, fill_value=0) # NaN values are replaced with )
```

```python
df1.rsub(df2, fill_value=0)
```

```python
df1.sub(df2)
```

```python
df2.rdiv(df1)
```

```python
df2.ediv(df1, fill_value=0) # data frame has no div
```
