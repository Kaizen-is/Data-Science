### Data Science and AI Practicum

- Pandasa(Dictionary)

```python
import pandas as pd
students = {
    "Alice": {"age": 20, "major": "Computer Science", "GPA": 3.8, "course": "Data Science"},
    "Bob": {"age": 22,"major": "Cybersecurity","GPA": 3.6, "course":"Ethical Hacking"}
}

df = pd.DataFrame(students)
df
```

```python
df.index.name = 'StudentINF'
df
```

```python
df = pd.DataFrame(students, index = ['age', 'GPA'])
df
```

```python
df = pd.DataFrame(students, index = ['age', 'GPA', 'major', 'course'])
df
```

```python
df.index.name = 'Students'
df.T
```
