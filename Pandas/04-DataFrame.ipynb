```python
import pandas as pd

ielts = {
    'Year':[2025, 2024, 2023, 2022, 2021 , 2020],
    'City':['Tashketn', 'Samarkand', 'Buhara', 'Jizzax', 'Andijon', 'Namangan'],
    'Average':[7, 6.5, 6.5, 7, 5, 5.5]
}
df = pd.DataFrame(ielts)
df
```

```python
df = pd.DataFrame(ielts, columns = ('City', 'Average', 'Year'))
df
```

```python
df = pd.DataFrame(ielts, columns = ('Year', 'City', 'Average', 'People'), index = (1, 2, 3, 4, 5, 6))
df
```
