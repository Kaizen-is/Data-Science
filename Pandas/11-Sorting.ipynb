### Data Science and AI Practicum

### Sorting

```python
import numpy as np
import pandas as pd

seri = pd.Series(np.arange(10), index = list('abcdefghij'))
seri
```

```python
seri.sort_values(ascending = False)
```

```python
seri.sort_index()
```

```python
seri.sort_index(ascending=False)
```

```python
df = pd.DataFrame(np.arange(12).reshape(4, 3), columns = list('abc'), index = ['Apple', 'Mango', 'Coconat', 'WatterMelon'])
df
```

```python
df.sort_index()
```

```python
df.sort_index(axis = 1, ascending=0)
```

```python
df.sort_index(axis=1, inplace=True)
df
```

```python
df.sort_index(axis = 1, inplace = False)
```

```python
df.sort_values(by = 'a')
```

```python
df.sort_values(by = 'a', ascending = False)
```

```python
df.sort_values(by = ['a', 'b', 'c'], ascending=False)
```

### Ranking

```python
df['Rating'] = df['a'].rank()
df
```

```python
ndf = df.sort_values(by = ['a', 'b', 'c'], ascending=False)
ndf
```

```python
ndf['Rating'] = ndf['c'].rank()
ndf
```

```python
ndf['Rating'] = ndf['c'].rank(method='min')
ndf
```

```python
ndf['Rating'] = ndf['a'].rank(method='average')
ndf
```
