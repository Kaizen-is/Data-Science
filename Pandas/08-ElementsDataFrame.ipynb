Choosing elements from DataFrame

```python
import pandas as pd

uz = {2020:35.4, 2021:36.1, 2022:37.2, 2023:40, 2024:42, 2015:44}
ka = {2020:32.4, 2021:33.1, 2022:34.2, 2023:40, 2024:32.2, 2015:34}
ta = {2020:44.4, 2021:42.7, 2022:45.3, 2023:46.3, 2024:48, 2015:50}
tu = {2020:63.4, 2021:65.1, 2022:70.2, 2023:70.5, 2024:72.3, 2015:77}

df = pd.DataFrame(
    {"Uzbekistan":uz,
     "Kazaxistan":ka,
     "Tadjikistan":ta,
     "Turkmenistan":tu}
)
df
```

```python
df['Uzbekistan'] # we get a column is Series data type
```

```python
df[['Uzbekistan']] # we get a DataFrame column
```

```python
type(df['Uzbekistan'])
```

```python
type(df[['Uzbekistan']])
```

```python
df['Uzbekistan'].to_numpy()
```

```python
df[['Uzbekistan']].to_numpy() # we easily can transfer dataframe or series to numpy array
```

```python
df[['Kazaxistan', 'Uzbekistan']] # we can get two columns only as dataframe
```

```python
df['Kazaxistan', 'Uzbekistan'] # two serieses can not be in one table as it is not a series, already a dataframe
```
