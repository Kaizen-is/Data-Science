### Data Science and AI Practicum

- Functions

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(3, 4), index = list('abc'), columns = ['Apple', 'Mango', 'Coconat', 'WatterMelon'])
```

```python
df
```

`abs` an absolute form of an element

```python
np.abs(df)
```

`Round` rounds elements

```python
np.round(df)
```

`Mean`  an average

```python
df.mean(axis = 0)
```

```python
df.mean(axis = 1)
```

`apply`  used to apply functions to dataframe

```python
df.apply(max)
```

```python
df.apply(sum)
```

```python
df.apply(min, axis = 1)
```

```python
df.apply(np.round) # or any other functions
```

`lambda` temporay function

```python
newfunc = lambda x: x.max() - x.min() # difference between max and min elements

df.apply(newfunc)
```

```python
dfint = pd.DataFrame(np.arange(9).reshape(3, 3), index = list('abc'), columns = ['Apple', 'Mango', 'Coconat'])
```

```python
dfint
```

```python
s = lambda x: x*x
dfint.apply(s)
```

```python
c = lambda x: x * 0.5
dfint.apply(c)
```

```python
def func(x):
  return pd.Series([x.max(), x.min()], index = ['max', 'min'])
dfint.apply(func)
```

`applymap` applies a function to each element

```python
dfint.applymap(lambda x: x * x)
```
