### DataStatistics: Correlation

```python
import pandas as pd
import numpy as np

user = {"age":[10, 20, 30, 40, 50, 60],
        "year":[2010, 2000, 1990, 1980, 1970, 1960]}

df = [pd.DataFrame(user)]
df
```

```python
df = pd.DataFrame(user, index = (1, 2, 3, 4, 5, 6))
df
```

`.corr()`

```python
df["age"].corr(df['year'])
```

```python
df.corr()
```

```python
uz = {2020:35.4, 2021:36.1, 2022:37.2, 2023:40, 2024:42, 2015:44}
ka = {2020:32.4, 2021:33.1, 2022:34.2, 2023:40, 2024:32.2, 2015:34}
ta = {2020:44.4, 2021:42.7, 2022:45.3, 2023:46.3, 2024:48, 2015:50}
tu = {2020:63.4, 2021:65.1, 2022:70.2, 2023:70.5, 2024:72.3, 2015:77}

df2 = pd.DataFrame(
    {"Uzbekistan":uz,
     "Kazaxistan":ka,
     "Tadjikistan":ta,
     "Turkmenistan":tu}
)
df2
```

```python
df2.corr()
```

```python
df2.T.corr()
```

We are getting a corralation between years and Countries

- We can get a corrallation between all or two columns or using `.corrwith()` we can get a corralation of one columns to all others

```python
df2.corrwith(df2["Uzbekistan"]) # we got corrolations through all other countries
```

As closer the result of corr to 0, no matter from whoch axis(-+), that means corralation is close, have similarities and this is a good result

```python
newdf = df2.T
```
