Filtering values using `.isin()`: returns false and true

```python
filter = [2019, 2018]
yfilter = df["year"].isin(filter)
df[yfilter]
```

```python
m_filter = ([" A Class", " SLK"])
filterm = df["model"].isin(m_filter)
filterm
```

```python
df[filterm] # getting columns appropriate to the filter
```

```python
df2 = df["model"].isin(m_filter)
df2
```

```python
new_filter = df["year"].isin([2019, 2018])
df[new_filter] # we get all the columns related to the filter
```

```python
df[df["year"]>200]
```

```python
m_filter = ([" A Class", " SLK"])

filterm = df["model"].isin(m_filter)
```

```python
df[df["model"].isin([" A Class", " SLK"])] # the same code as above but in more simple way
```

Getting cars only related to `A Class and SLK` and the only ones that are newer than 2000 and working on Petrol and Diesel

```python
df[(df["model"].isin([" A Class", " SLK"])) & (df["year"]>2000) & df["fuelType"].isin(["Petrol", "Diesel"])]
```
