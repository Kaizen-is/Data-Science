###`.map()`

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data_nan.csv', index_col = 'index')
df
```

```python
df['num-of-cylinders'].unique()
```

```python
dictt = {'four':4, 'six':6, 'five':5, 'twelve':12, 'two':2, 'eight':8, 'three':3}
df['cylinders'] = df['num-of-cylinders'].map(dictt)
df
```

```python
uzs = 12700
df['price_uzs'] = df['price'].map(lambda x: x * uzs)
df['price_uzs'] = df['price_uzs'].fillna(np.mean(df['price_uzs']))
df['price_uzs'] = df['price_uzs'].astype(np.int32)
df.head()
```

###`.replace()`

```python
df.replace({'six':6})
```

```python
df.replace(dictt)
```

### `.cut()`

```python
df['Population'] = df['Population'].astype(np.int16)
df.head()
```

```python
df.describe()
```

```python
bins = [0, 1000, 4000]
groups = pd.cut(df['Population'], bins)
groups.value_counts()
```

### `.groupby()`

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data.csv', index_col = 'index')
df
```

```python
df.groupby(['company', 'body-style']).min()
```

```python
df[(df['company'] == 'alfa-romero') & (df['body-style'] == 'convertible')]['horsepower'].min()
```

```python
df.groupby('company')['price'].mean()
```

### `.agg([])`

```python
df.groupby('company')[['horsepower', 'average-mileage', 'price', 'wheel-base', 'length']].agg([min, max])
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/MOCK_USER_DATA.csv', index_col = 'id')
df.head()
```

We can note that in age column sth is going wrong. Max age of the person in history is 123 and min 1 so - 85 is wrong

```python
df.describe()
```

```python
df['age'] = df['age'].apply(abs)
indx = df[df['age'] > 123].index
indx
```

```python
df.drop(index = indx, inplace = True)
df.head()
```

Now we can see max and min age values are appropriate

```python
df.describe()
```

But now gender column has variety types, we only need `Male` and `Female`

```python
df['gender'].value_counts()
```

```python
df = df[df['gender'].isin(['Male', 'Female'])]
df.head()
```

```python
df.columns = df.columns.str.title()
df['Gender'].value_counts()
```
