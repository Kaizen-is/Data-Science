### Working with Dates

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("database.csv.zip")
df
```

```python
df.columns
```

```python
cols = ['Date', 'Time', 'Latitude', 'Longitude', 'Type', 'Depth', 'DepthErr', 'DepSeismic', 'Magnitude', 'MagTypes',
        'MagErr', 'MagSeismic', 'AzimuthalGap', 'HorizDist', 'HorizErr', 'RootMean', 'ID', 'Source', 'LocSource', 'MagSource', 'Status']
df.columns = cols
df
```

```python
df.info()
```

```python
df.DepthErr = df.DepthErr.fillna(np.mean(df.DepthErr))
df.DepSeismic = df.DepSeismic.fillna(np.mean(df.DepSeismic))
df.MagSeismic = df.MagSeismic.fillna(np.mean(df.MagSeismic))
df.AzimuthalGap = df.AzimuthalGap.fillna(np.mean(df.AzimuthalGap))
df.HorizDist = df.HorizDist.fillna(np.mean(df.HorizDist))
df.HorizErr = df.HorizErr.fillna(np.mean(df.HorizErr))
df.RootMean = df.RootMean.fillna(np.mean(df.RootMean))
df.MagErr = df.MagErr.fillna(np.mean(df.MagErr))

df.Latitude = round(df.Latitude, 2)
df.Longitude = round(df.Longitude, 2)
df.DepthErr = round(df.DepthErr, 3)
df.HorizDist = round(df.HorizDist, 2)
df.RootMean = round(df.RootMean, 2)
df.MagSeismic = round(df.MagSeismic, 2)
df.AzimuthalGap = round(df.AzimuthalGap, 2)
df.HorizErr = round(df.HorizErr, 2)
df.DepSeismic = round(df.DepSeismic, 3)
df.MagErr = round(df.MagErr, 2)

df.index.name = 'Index'
```
