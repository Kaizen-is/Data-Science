### Inserting tables in to a DataBase using pandas

```python
import pandas as pd
import sqlite3 as sql
import numpy as np
```

```python
db = sql.connect('Auto.db')
cursor = db.cursor()
```

Reading CSV

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data.csv', index_col = 'index')
df
```

```python
cols = ['company', 'body', 'wheel', 'length', 'engine', 'cylinders', 'horse', 'milage', 'price']
df.columns = cols
df.head()
```

A bit manipulation

```python
df['price'] = df['price'].replace({np.nan:0})
df['price'] = df['price'].astype(np.int16)
df
```

```python
df.to_sql('auto_info', db)
```

### Data Exchanging

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/world_population.csv', index_col = 'rank')
df
```

```python
uz = df[df['country'] == 'Uzbekistan']
uz
```

```python
rank = uz.index.values[0]
rank
```

```python
country = uz.at[rank, 'country']
density = uz.at[rank, 'density']
densityMi = uz.at[rank, 'densityMi']
pop = uz.at[rank, 'pop2021']
area = uz.at[rank, 'area']
```

Creating another empty table

```python
query = """
create table WorldPop(
  country varchar(100),
  density real,
  densityMi real,
  pop real,
  area integer
)
"""
```

```python
cursor.execute(query)
db.commit()
```

```python
cursor.execute('select name from sqlite_master where type = "table"')
cursor.fetchall()
```

```python
cursor.execute('pragma table_info(WorldPop)')
cursor.fetchall()
```

Insertion process

```python
query = f"""
insert into worldPop(country, density, densityMi, pop, area)
values('{country}', {density}, {densityMi}, {pop}, {area})
"""
```

```python
cursor.execute(query)
db.commit()
```

We have inseted an information from csv to our table in db, which is ususally used to make process of insertion automatic

```python
query = 'select * from WorldPop'
pd.read_sql(query, db)
```
