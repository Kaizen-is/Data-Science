###  Fill NaN

`.fillna()`

```python
df3 = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data_nan.csv', index_col = 'index')
df3.head(9)
```

`.fillna()`

```python
df3.fillna(0)
```

We have replaced all nan values in the wheel column with the mean of this col and changed dt into int16

```python
df3['wheel-base'] = df3['wheel-base'].fillna(np.mean(df3['wheel-base']))
df3['wheel-base'] = df3['wheel-base'].astype(np.int16)
df3
```

```python
df3 = df3.fillna({'horsepower':100, 'average-mileage':200})
df3
```

```python
df2
```

```python
df2 = df2.fillna({'horsepower':np.mean(df2['horsepower']), 'average-mileage':np.mean(df2['average-mileage'])})
df2['horsepower'] = round(df2['horsepower'], 1)
df2['average-mileage'] = round(df2['average-mileage'], 1)
df2.head(9)
```

`.fillna(methed = ffill|bfill)`

```python
df2['wheel-base'] = df2['wheel-base'].fillna(method = 'ffill')
df2
```

### According to the codion

```python
df2 = df2[df2['num-of-cylinders'] == 'four'].fillna(method = 'ffill')
df2
```

```python
coll = ['company', 'style', 'wheel', 'length', 'engine', 'cylinders', 'horsepower', 'milage', 'price']
df2.columns = coll
df2['wheel'] = df2['wheel'].astype(np.int16)
df2['length'] = df2['length'].astype(np.int16)
df2['price'] = df2['price'].astype(np.int32)
df2['horsepower'] = df2['horsepower'].astype(np.int16)
df2.head(5)
```
