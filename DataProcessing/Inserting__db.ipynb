### Inserting tables in to a DataBase using pandas

```python
import pandas as pd
import sqlite3 as sql
import numpy as np
```

```python
db = sql.connect('Auto.db')
cursor = db.cursor()
```

Reading CSV

```python
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/automobile_data.csv', index_col = 'index')
df
```

```python
cols = ['company', 'body', 'wheel', 'length', 'engine', 'cylinders', 'horse', 'milage', 'price']
df.columns = cols
df.head()
```

A bit manipulation

```python
df['price'] = df['price'].replace({np.nan:0})
df['price'] = df['price'].astype(np.int16)
df
```

```python
df.to_sql('auto_info', db)
```

Another db

```python
df2 = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/refs/heads/main/car_details.csv', index_col = 'name')
df2
```

```python
col = ['year', 'sell', 'km', 'fuel', 'seller', 'transmission', 'owner', 'mileage', 'engine', 'power', 'torque', 'seats']
df2.columns = col
df2
```

### Replace nan values with 0 to implement some manipulations

```python
df2['mileage'] = df2['mileage'].replace({np.nan:0})
df2['mileage'] = df2['mileage'].astype(np.int16)
df2.head(5)
```

```python
df2['engine'] = df2['engine'].replace({np.nan:0})
df2['engine'] = df2['engine'].astype(np.int32)
df2.head(5)
```

```python
df2['power'] = df2['power'].replace({np.nan:0})
df2['power'] = df2['power'].astype(np.int16)
df2
```

```python
df2['seats'] = df2['seats'].replace({np.nan:0})
df2['seats'] = df2['seats'].astype(np.int16)
df2
```

```python
df2.to_sql('CarDetails.db', db)
```

### Checking Tables in our db

```python
query = "select name from sqlite_master where type = 'table';"
```

```python
cursor.execute(query)
cursor.fetchall()
```

Changing the name of the table that was already saved

```python
cursor.execute("ALTER TABLE 'CarDetails.db' RENAME TO CarDetails;")
db.commit()
```

```python
query = "select name from sqlite_master where type = 'table';"
cursor.execute(query)
cursor.fetchall()
```

`pragma table_info`

```python
cursor.execute('pragma table_info(auto_info);')
cursor.fetchall()
```
