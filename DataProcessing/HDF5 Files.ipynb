### Pandas: HDF5 Files

```python
import pandas as pd
```

```python
trains = pd.read_csv("sample_data/mnist_train_small.csv")
trains
```

```python
comb = pd.read_json("sample_data/anscombe.json")
comb
```

```python
hdf = pd.HDFStore("Data.h5")
```

```python
hdf.put("train", trains)
hdf.put("combe", comb)
```
