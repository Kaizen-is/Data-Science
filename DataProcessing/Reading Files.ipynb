- By default, pandas sets a new index column. We can change or just set new one

`index_col` used to set a column as an index col

```python
df_c = pd.read_csv('sample_data/california_housing_test.csv', index_col = 0) # or just write a name of a column
df_c
```

`names` used to set a new name to the existing columns of a table

```python
df_c = pd.read_csv('sample_data/california_housing_test.csv', index_col = 0, names = [
    'Long', 'Latitude', 'MedianAge', 'RoomsCount', 'BedroomsC', 'Popul', 'Rents', 'MedianIncome', 'MedianValueHouse'
    ]) # or just write a name of a column
df_c
```
