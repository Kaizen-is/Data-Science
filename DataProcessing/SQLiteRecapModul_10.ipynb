````python
import requests
````

````python
coins_url = "https://api.coingecko.com/api/v3/coins/list"
api = requests.get(coins_url)
df_api = api.json()
df_api
````

````python
df_api = pd.DataFrame(df_api)
df_api
````

Searching for Bitcoin

````python
df_api[df_api['id'] == 'bitcoin']
````

````python
df_api.head()
````

````python
coin_ids = df_api[df_api['id'].str.match(r'^[a-zA-Z0-9]+$')]['id'].tolist()
coin_ids = coin_ids[-10:]
coin_ids
````

#**6. Working with APIS**

https://api.coingecko.com/api/v3/simple/price?ids={coin_ids}&vs_currencies=usd


````python
hdf_coin = pd.HDFStore('CryptoCurency.h5')
for i in coin_ids:
  price_url = f"https://api.coingecko.com/api/v3/simple/price?ids={i}&vs_currencies=usd"
  a = requests.get(price_url)
  b = a.json()
  df = pd.DataFrame(b).T
  df.index.name = 'name'
  hdf_coin.put(f'{i}', df)

hdf_coin.keys()
#hdf_coin.close()
````
