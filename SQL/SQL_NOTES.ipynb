### Joins

- Inner Join: takes two or more tables and creates a new table(view) with appropriate info only

image.png

`Left Join`: Creates a new tabel(view) only with the values that are suitable  with the second tables values and its self

image.png

```python
SELECT column1, column2, ...
FROM table1
LEFT JOIN table2
ON table1.column = table2.column;
```

`Right Join`: Anological to left join

image.png

```python
SELECT column1, column2, ...
FROM table1
RIGHT JOIN table2
ON table1.column = table2.column;
```

`Full Join`: Creates a table taking all the values, appropriate ones and not

image.png

```python
SELECT column1, column2, ...
FROM table1
FULL OUTER JOIN table2
ON table1.column = table2.column;
```

`Cross Join`: Returns all combinations of the rows(Kartez)

```python
SELECT column1, column2, ...
FROM table1
CROSS JOIN table2;
```

`Self Join`: Joins a tables with it self

```python
SELECT A.name AS Employee, B.name AS Manager
FROM employees A
JOIN employees B ON A.manager_id = B.employee_id;
```

### Unions

`Union`: joins row by row an appropriate values from two queries

image.png

```python
SELECT FirstName FROM Employees
UNION
SELECT FirstName FROM Customers;
```

`Union All`: Particularly the same with Union, just gets duplicates also

image.png

```python
SELECT FirstName FROM Employees
UNION ALL
SELECT FirstName FROM Customers;
```

`INTERSECT`: Gets only only appropriate ones

image.png

```python
SELECT FirstName FROM Employees
INTERSECT
SELECT FirstName FROM Customers;
```

`EXCEPT`: Gets values from first query and does not get values that consists in the second query

image.png

```python
SELECT FirstName FROM Employees
EXCEPT
SELECT FirstName FROM Customers;
```

**Examples**

```python

SELECT p.FirstName
FROM Person.Person p
JOIN HumanResources.Employee e ON p.BusinessEntityID = e.BusinessEntityID

UNION

SELECT p.FirstName
FROM Person.Person p
JOIN Sales.Customer c ON p.BusinessEntityID = c.PersonID;
```

### String Functions

`1.LEN(sting)`: Mesures length of the sring value

```python
SELECT LEN('Hello World')  -- Result: 11
```

`2.DATALENGTH(string)`: Same with len but spaces are taken in to account while len does not

```python

SELECT DATALENGTH('Hello ')  -- Result: 6
# sace is counted
```

`3.LEFT(string, length)`: From first index to 9, letters will be taken

```python
SELECT LEFT('AdventureWorks', 9)  --Result: Adventure
```

`4.RGIHT(sting, length)`: from right side 5 indexes are takes

```python
SELECT LEFT('AdventureWorks', 9)  -- Result: Adventure
```

`5.CHARINDEX(substring, string)`: Gives an index from where substring is started

```python
SELECT CHARINDEX('W', 'AdventureWorks')  -- Result: 10
```

`6.PATINDEX(%pattern%, string)`: The same with like, gives an index from where pattern is started

```python
SELECT PATINDEX('%Work%', 'AdventureWorks')  -- Result: 10
```

`7.REPLACE(sting, old_s, new_s)`: Replaces old word to a new one

```python
SELECT REPLACE('I like SQL', 'like', 'love')  -- Result: I love SQL
```

`8.LTRIM(string)`: Gets rid of extra spaces from left

```python
SELECT LTRIM('   Hello')  -- Result: 'Hello'
```

`9.RTRIM(string)`: Gets rid of extra spaces from right

```python
SELECT RTRIM('Hello   ')  -- Result: 'Hello'
```

`10.LOWER(string)`: All letters will be turned into lower case letters

```python
SELECT LOWER('SQL SERVER')  -- Result: sql server
```

`11.UPPER(string)`: All letters will be turned into upper case letters

```python
SELECT UPPER('sql server')  -- Result: SQL SERVER
```

`12.CONCAT(s1, s2)`: Concatenates two strings

```python
SELECT CONCAT('Hello', ' ', 'World')  -- Result: Hello World
```

`13.FORMAT(value, format)`: formates the value in to exact data type

```python
SELECT FORMAT(12345.6789, 'N2')  -- Result: 12,345.68
```

`14.REPLICATE(string, times)`: replicates string many times as given

```python
SELECT REPLICATE('ABC', 3)  -- Result: ABCABCABC
```

`15.REVERSE(string)`: string given will be writen oppositely

```python
SELECT REVERSE('SQL')  -- Result: LQS
```

`16.Space(times)`: times - creates a space

```python
SELECT 'Hello' + SPACE(3) + 'World'  -- Result: Hello   World
```

`17.CAST(expression as data_type)`: changes one data type to another

```python

SELECT CAST(123 AS VARCHAR(10));  -- Result: '123'

SELECT CAST(GETDATE() AS VARCHAR(30));  -- Result: 'Mar 26 2025 11:45AM'
```

`18.CONVERT(dtype, expression, style)`: fast the same with cast, just time and date will be given more precisely

```python
-- GETDATE()(style = 104 — German way: dd.mm.yyyy)
SELECT CONVERT(VARCHAR, GETDATE(), 104);
-- '26.03.2025'

(style = 23)
SELECT CONVERT(VARCHAR, GETDATE(), 23);
-- '2025-03-26'
```

`19.ASCII(character)`: gives an ascii value of the char

```python
SELECT ASCII('A');  -- Result: 65
```

`20.CHAR(ascii_code)`: wise versa to ascii

```python
SELECT CHAR(65);  -- Result: 'A'
```
